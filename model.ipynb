```python
mid = 9
extra = ""

activation = "linear"
initializer = "glorot_uniform"

inputs = Input(shape=(512,512,1))

x = ZeroPadding2D((1,1))(inputs)
x = Convolution2D(64,(3,3), kernel_initializer=initializer, activation=activation)(x)
x = ZeroPadding2D((1,1))(x)
x = Convolution2D(64,(3,3), kernel_initializer=initializer, activation=activation)(x)
x = MaxPooling2D((2,2),strides=(2,2))(x)

x = ZeroPadding2D((1,1))(x)
x = Convolution2D(64,(3,3), kernel_initializer=initializer, activation=activation)(x)
x = ZeroPadding2D((1,1))(x)
x = Convolution2D(64,(3,3), kernel_initializer=initializer, activation=activation)(x)
x = MaxPooling2D((2,2),strides=(2,2))(x)

x = ZeroPadding2D((1,1))(x)
x = Convolution2D(64,(3,3), kernel_initializer=initializer, activation=activation)(x)
x = ZeroPadding2D((1,1))(x)
x = Convolution2D(64,(3,3), kernel_initializer=initializer, activation=activation)(x)
x = MaxPooling2D((2,2),strides=(2,2))(x)

x = ZeroPadding2D((1,1))(x)
x = Convolution2D(128,(3,3), kernel_initializer=initializer, activation=activation)(x)
x = ZeroPadding2D((1,1))(x)
x = Convolution2D(128,(3,3), kernel_initializer=initializer, activation=activation)(x)
x = MaxPooling2D((2,2),strides=(2,2))(x)

x = ZeroPadding2D((1,1))(x)
x = Convolution2D(128,(3,3), kernel_initializer=initializer, activation=activation)(x)
x = ZeroPadding2D((1,1))(x)
x = Convolution2D(128,(3,3), kernel_initializer=initializer, activation=activation)(x)
x = MaxPooling2D((2,2),strides=(2,2))(x)

x = ZeroPadding2D((1,1))(x)
x = Convolution2D(128,(3,3), kernel_initializer=initializer, activation=activation)(x)
x = ZeroPadding2D((1,1))(x)
x = Convolution2D(128,(3,3), kernel_initializer=initializer, activation=activation)(x)
x = MaxPooling2D((2,2),strides=(2,2))(x)

x = ZeroPadding2D((1,1))(x)
x = Convolution2D(256,(3,3), kernel_initializer=initializer, activation=activation)(x)
x = ZeroPadding2D((1,1))(x)
x = Convolution2D(256,(3,3), kernel_initializer=initializer, activation=activation)(x)
x = MaxPooling2D((2,2),strides=(2,2))(x)

x = ZeroPadding2D((1,1))(x)
x = Convolution2D(256,(3,3), kernel_initializer=initializer, activation=activation)(x)
x = ZeroPadding2D((1,1))(x)
x = Convolution2D(256,(3,3), kernel_initializer=initializer, activation=activation)(x)
x = MaxPooling2D((2,2),strides=(2,2))(x)

x = ZeroPadding2D((1,1))(x)
x = Convolution2D(256,(3,3), kernel_initializer=initializer, activation=activation)(x)
x = ZeroPadding2D((1,1))(x)
x = Convolution2D(256,(3,3), kernel_initializer=initializer, activation=activation)(x)
x = MaxPooling2D((2,2),strides=(2,2))(x)

x = Flatten()(x)
x = Dense(1024, kernel_initializer=initializer, activation=activation)(x)
x = Dropout(0.5)(x)
x = Dense(1024, kernel_initializer=initializer, activation=activation)(x)
x = Dropout(0.5)(x)
outputs = Dense(2,activation="softmax")(x)

model = Model(inputs = inputs, outputs=outputs)
model.summary()

optimizer = Adam(lr=0.0001)

model.compile(optimizer=optimizer, loss="binary_crossentropy", metrics=["accuracy"])
```

```python
from keras.preprocessing.image import ImageDataGenerator

batch_size = 8

datagen = ImageDataGenerator(rescale=1./255)
train_generator = datagen.flow_from_directory(
    'data/train',
    target_size=(512, 512),
    batch_size=batch_size,
    class_mode='categorical',
    color_mode='grayscale'
)

validate_generator = datagen.flow_from_directory(
    'data/validate',
    target_size=(512, 512),
    batch_size=batch_size,
    class_mode='categorical',
    color_mode='grayscale'
)

#test_generator = datagen.flow_from_directory(
#    'data/test',
#    target_size=(512, 512),
#    batch_size=5,
#    class_mode='categorical'
#)
```

```python
import os

from keras.callbacks import TensorBoard, ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from math import ceil

from datetime import datetime
from time import time


ts = datetime.fromtimestamp(time()).strftime('%Y-%m-%d-%H-%M-%S')

os.makedirs("./models/MID{mid:03d}[{extra}]_{ts}".format(mid=mid,ts=ts,extra=extra))

tensorboard = TensorBoard(log_dir="./tensorboard/MID{mid:03d}[{extra}]_{ts}".format(mid=mid,ts=ts,extra=extra))
checkpoint = ModelCheckpoint("./models/MID{mid:03d}[{extra}]_{ts}/weights.{{epoch:02d}}-{{val_loss:.2f}}.hdf5".format(mid=mid,ts=ts,extra=extra))
lrchange = ReduceLROnPlateau(monitor='val_acc', factor=0.5, patience=3, min_lr=0.00001, verbose=1)
stopper = EarlyStopping(monitor='val_acc', min_delta=0, patience=6, verbose=1)

model.fit_generator(
    train_generator, 
    steps_per_epoch=ceil(27324/batch_size), 
    epochs=50,
    validation_data=validate_generator, 
    validation_steps=ceil(3917/batch_size),
    callbacks=[tensorboard, checkpoint, stopper, lrchange]
)
```
